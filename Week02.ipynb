# Câu 1 - Đáp án A

```python
def max_kernel(num_list, k):
    result = []
    n = len(num_list)

    for i in range(n - k + 1):
        # Find the maximum in the current window
        max_in_window = max(num_list[i:i + k])
        result.append(max_in_window)

    return result

# Test case
assert max_kernel([3, 4, 5, 1, -44], 3) == [5, 5, 5]
num_list = [3, 4, 5, 1, -44, 5, 10, 12, 33, 1]
k = 3
print(max_kernel(num_list, k))
```

# Câu 2 - Đáp án A

```python
def character_count(word):
    character_statistic = {}

    for char in word:
        if char != ' ':
            if char in character_statistic:
                character_statistic[char] += 1
            else:
                character_statistic[char] = 1

    return character_statistic

# Test case
assert character_count("Baby") == {'B': 1, 'a': 1, 'b': 1, 'y': 1}
print(character_count('smiles'))
```

```python
def count_word(file_path):
    counter = {}

    # Open and read the file
    with open(file_path, 'r') as file:
        for line in file:
            words = line.split()
            for word in words:
                word = word.lower()  # Convert to lowercase
                if word in counter:
                    counter[word] += 1
                else:
                    counter[word] = 1

    return counter

result = count_word(file_path)
assert result['who'] == 3
print(result['man'])
```

# Câu 4 - Đáp án A

```python
def levenshtein_distance(token1, token2):
    m = len(token1) + 1
    n = len(token2) + 1

    # Initialize matrix
    D = [[0] * n for _ in range(m)]

    # Initialize the first row and first column
    for i in range(m):
        D[i][0] = i
    for j in range(n):
        D[0][j] = j

    # Populate the matrix
    for i in range(1, m):
        for j in range(1, n):
            cost = 0 if token1[i - 1] == token2[j - 1] else 1
            D[i][j] = min(D[i - 1][j] + 1,  # Deletion
                          D[i][j - 1] + 1,  # Insertion
                          D[i - 1][j - 1] + cost)  # Substitution

    return D[m - 1][n - 1]

# Test cases
assert levenshtein_distance("hi", "hello") == 4.0
print(levenshtein_distance("hola", "hello"))
```

# Câu 5 - Đáp án A

```python
def check_the_number(N):
    list_of_numbers = []
    results = ""
    for i in range(1, 5):
        # Sử dụng append thêm i vào list_of_numbers
        list_of_numbers.append(i)

    if N in list_of_numbers:
        results = "True"
    if N not in list_of_numbers:
        results = "False"

    return results

N = 7
assert check_the_number(N) == "False"

N = 2
results = check_the_number(N)
print(results)
```

# Câu 6 - Đáp án C

```python
def my_function(data, max, min):
    result = []
    for i in data:
        if i < min:
            result.append(min)
        elif i > max:
            result.append(max)
        else:
            result.append(i)
    return result


my_list = [5, 2, 5, 0, 1]
max_value = 1
min_value = 0
assert my_function(data=my_list, max=max_value, min=min_value) == [1, 1, 1, 0, 1]


my_list = [10, 2, 5, 0, 1]
max_value = 2
min_value = 1
print(my_function(data=my_list, max=max_value, min=min_value))
```

# Câu 10 - Đáp án C

```python
def My_function(integers, number=1):
    comparison_list = [i == number for i in integers]
    return any(comparison_list)

my_list = [1, 3, 9, 4]
assert My_function(my_list, -1) == False

my_list = [1, 2, 3, 4]
print(My_function(my_list, 2))
```

# Câu 13 - Đáp án C

```python
def my_function(y):
    var = 1
    while y > 1:
        var *= y
        y -= 1
    return var

assert my_function(8) == 40320
print(my_function(4))
```

# Câu 15 - Đáp án C

```python
def function_helper(x):
    if x > 0:
        return 'T'
    else:
        return 'N'

def my_function(data):
    res = [function_helper(x) for x in data]
    return res

data = [10, 0, -10, -1]
assert my_function(data) == ['T', 'N', 'N', 'N']

data = [2, 3, 5, -1]
print(my_function(data))
```

# Câu 16 - Đáp án A

```python
def function_helper(x, data):
    for i in data:
        if x == i:
            return 0
    return 1

def my_function(data):
    res = []
    for i in data:
        if function_helper(i, res):
            res.append(i)
    return res

lst = [10, 10, 9, 7, 7]
assert my_function(lst) == [10, 9, 7]

lst = [9, 9, 8, 1, 1]
print(my_function(lst))
```
