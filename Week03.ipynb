# Cu 2 - B

```python
class MySoftmax(nn.Module):
    def __init__(self):
        super().__init__()

    def forward(self, x):
        exp_x = torch.exp(x)
        sum_exp_x = torch.sum(exp_x)
        softmax_x = exp_x / sum_exp_x
        return softmax_x

data = torch.Tensor([5, 2, 4])
my_softmax = MySoftmax()
output = my_softmax(data)
print(output)  # For verification

assert round(output[-1].item(), 2) == 0.26
output
```

# Cu 3 - C

```python
class MySoftmax(nn.Module):
    def __init__(self):
        super().__init__()

    def forward(self, x):
        c = torch.max(x)
        exp_x = torch.exp(x - c)
        sum_exp_x = torch.sum(exp_x)
        softmax_x = exp_x / sum_exp_x
        return softmax_x

data = torch.Tensor([1, 2, 300000000])
my_softmax = MySoftmax()
output = my_softmax(data)
print(output)  # For verification

assert round(output[0].item(), 2) == 0.0
output
```

# Cu 4 - B

```python
class SoftmaxStable(nn.Module):
    def __init__(self):
        super().__init__()

    def forward(self, x):
        x_max = torch.max(x)
        x_exp = torch.exp(x - x_max)
        partition = x_exp.sum(0)
        return x_exp / partition

data = torch.Tensor([1, 2, 3])
softmax_stable = SoftmaxStable()
output = softmax_stable(data)
print(output)  # For verification

assert round(output[-1].item(), 2) == 0.67
output
```

# Cu 8 - C

```python
class Person(ABC):
    def __init__(self, name: str, yob: int):
        self._name = name
        self._yob = yob
    
    def get_yob(self):
        return self._yob

    @abstractmethod
    def describe(self):
        pass

class Student(Person):
    def __init__(self, name: str, yob: int, grade: str):
        super().__init__(name, yob)
        self._grade = grade
    
    def describe(self):
        print(f"Student - Name: {self._name} - YoB: {self._yob} - Grade: {self._grade}")

class Teacher(Person):
    def __init__(self, name: str, yob: int, subject: str):
        super().__init__(name, yob)
        self._subject = subject
    
    def describe(self):
        print(f"Teacher - Name: {self._name} - YoB: {self._yob} - Subject: {self._subject}")

class Doctor(Person):
    def __init__(self, name: str, yob: int, specialist: str):
        super().__init__(name, yob)
        self._specialist = specialist
    
    def describe(self):
        print(f"Doctor - Name: {self._name} - YoB: {self._yob} - Specialist: {self._specialist}")

class Ward:
    def __init__(self, name: str):
        self.__name = name
        self.__listPeople = list()
    
    def add_person(self, person: Person):
        self.__listPeople.append(person)
    
    def describe(self):
        print(f"Ward Name: {self.__name}")
        for p in self.__listPeople:
            p.describe()
    
    def count_doctor(self):
        count = 0
        for person in self.__listPeople:
            if isinstance(person, Doctor):
                count += 1
        return count

# To cc i tng
student1 = Student(name="studentA", yob=2010, grade="7")
teacher1 = Teacher(name="teacherA", yob=1969, subject="Math")
teacher2 = Teacher(name="teacherB", yob=1995, subject="History")
doctor1 = Doctor(name="doctorA", yob=1945, specialist="Endocrinologists")
doctor2 = Doctor(name="doctorB", yob=1975, specialist="Cardiologists")

# To mt Ward
ward1 = Ward(name="Ward1")
ward1.add_person(student1)
ward1.add_person(teacher1)
ward1.add_person(teacher2)
ward1.add_person(doctor1)
ward1.add_person(doctor2)

# In thng tin Ward v m s lng doctor
ward1.describe()
print(f"Number of doctors: {ward1.count_doctor()}")
```

```python
class MyStack:
    def __init__(self, capacity):
        self.__capacity = capacity
        self.__stack = []

    def is_full(self):
        return len(self.__stack) == self.__capacity

    def push(self, value):
        if not self.is_full():
            self.__stack.append(value)
        else:
            raise Exception("Stack is full")

# Khi to stack vi capacity = 5
stack1 = MyStack(capacity=5)

# Thm phn t vo stack
stack1.push(1)
assert stack1.is_full() == False  # Stack cha y
stack1.push(2)

# Kim tra trng thi y ca stack
print(stack1.is_full())
```

# Cu 10 - B

```python
class MyStack:
    def __init__(self, capacity):
        self.__capacity = capacity
        self.__stack = []

    def is_full(self):
        return len(self.__stack) == self.__capacity

    def push(self, value):
        if not self.is_full():
            self.__stack.append(value)
        else:
            raise Exception("Stack is full")

    def top(self):
        if len(self.__stack) == 0:
            return None
        return self.__stack[-1]

# Khi to stack vi capacity = 5
stack1 = MyStack(capacity=5)

# Thm phn t vo stack
stack1.push(1)
assert stack1.is_full() == False  # Stack cha y
stack1.push(2)

# Kim tra gi tr top ca stack
print(stack1.top())
```

# Cu 12 - D

```python
class MyQueue:
    def __init__(self, capacity):
        self.__capacity = capacity
        self.__queue = []

    def isEmpty(self):
        return len(self.__queue) == 0

    def is_full(self):
        return len(self.__queue) == self.__capacity

    def dequeue(self):
        if not self.isEmpty():
            return self.__queue.pop(0)
        else:
            raise Exception("Queue is empty")

    def enqueue(self, value):
        if not self.is_full():
            self.__queue.append(value)
        else:
            raise Exception("Queue is full")

    def front(self):
        if not self.isEmpty():
            return self.__queue[0]
        else:
            raise Exception("Queue is empty")

# Khi to queue vi capacity = 5
queue1 = MyQueue(capacity=5)

# Thm phn t vo queue
queue1.enqueue(1)
assert queue1.is_full() == False  # Queue cha y
queue1.enqueue(2)

# Ly phn t u tin trong queue
print(queue1.front())
```
